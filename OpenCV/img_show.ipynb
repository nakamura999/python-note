### 画像をインライン表示する

```python
%matplotlib inline
```

```python
# 画像を画面に表示
import matplotlib.pyplot as plt
import cv2
img = cv2.imread("trees-style32.jpg")

# 画像を出力 imshow()  色変換（RGBに） cvtColor
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show
```

```python
plt.axis("off") # 数値のメモリーを表示しない
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show
```

### 画像をpngファイルで保存

```python
cv2.imwrite("style.png", img)
```

### 画像をリサイズ

```python
# import matplotlib.pyplot as plt
# import cv2

# 画像をリサイズ
im2 = cv2.resize(img, (600,300))

# 画像を保存
cv2.imwrite("style-risize.png", im2)

# 画像を表示
plt.imshow(cv2.cvtColor(im2, cv2.COLOR_BGR2RGB))
plt.show()
```

### 画像の切り取り

```python
# 画像の一部を切り取り
im2 = img[150:750, 250:1000]

# 画像をリサイズ
im2 = cv2.resize(im2, (400,400))

# 保存
cv2.imwrite("style-cut.png", im2)

# 画像を表示
plt.imshow(cv2.cvtColor(im2, cv2.COLOR_BGR2RGB))
plt.show()
```
